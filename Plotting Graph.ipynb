```python
import matplotlib.pyplot as plt
```

# Line 

```python
x=[1,2,3]
y=[2,4,6]
#scatter the point all over the graph
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
#first character of string represent color and second character represent marker
plt.plot(x,y,"b--")
plt.show()
```

```python
#plot function x^3
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y)
plt.show()
```

```python
#color,marker,linewidth,label for axis,title
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="blue",marker="o",linewidth=4)
plt.xlabel("x")
plt.ylabel("x^3")
plt.title("line graph")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y1=x**3
y2=x*2
plt.plot(x,y1,color="black",linewidth=4,label="x^3")
plt.plot(x,y2,color="blue",linewidth=4,label="x^2")
plt.ylabel("y")
plt.xlabel("x")
plt.axis([0,5,0,100])
plt.text(4.2,60,"x^3")
plt.text(4,15,"x^2")
plt.legend()#for showing label
plt.title("x^2 vs x^3")
plt.grid()
plt.show()
```

# Bubble Chart
- size of bubble will tell how many people working at that time

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,16,18,20,23]
population=[100,200,180,250,300,370]
```

```python
plt.scatter(year,salary,s=90)
plt.show()
```

```python
plt.scatter(year,salary,s=population,alpha=0.8)
plt.title("Bubble chart")
plt.show()
```

```python
import numpy as np
c=np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.9,edgecolor="black")
plt.show()
```

# Pie Graph

```python
size=[3,6,5,2]
colors=["red","blue","green","black"]
labels=["A","B","C","D"]
plt.title("split among classes")
explode=[0.2,0,0,0]
plt.axis("equal")
plt.pie(size,colors=colors,labels=labels,explode=explode,autopct="%.2f%%")
plt.show()
```

```python
plt.pie(size,colors=colors,labels=labels,explode=explode,autopct="%.2f%%",counterclock=False,startangle=120)
plt.show()
```

# Histogram

```python
a=[1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(a)
plt.show()
```

```python
import numpy as np
xt=np.arange(len(a)+1)
plt.xticks(xt)
plt.hist(a,edgecolor="black")
plt.show()
```

```python
plt.hist(a,bins=21,edgecolor="black")
plt.xticks(xt)
plt.show()
```

# Bar Graph
